```python
root = '/Users/yangqingwen/Desktop/Github/2022Fintech-project'
testpath = f'{root}/predictions/0527-1104_test_pred.csv'
```

```python
import pandas as pd
testdf = pd.read_csv(testpath)
testdf.head()
```

```python
testdf['pclass_num'] = testdf['product'].apply(lambda x: len(x.split(';')))
```

```python
testdf['pclass_num'].value_counts()
```

```python
import json 
catDpath = f'{root}/data/category/category_dictionary.json'
with open(catDpath, 'r') as f:
    catD = json.load(f)
categories = list(catD.keys())
```

```python
from collections import defaultdict
combD = defaultdict(list)
testcatD = defaultdict(list, { k:[] for k in categories})
for i in range(len(testdf)):
    row = testdf.iloc[i]
    pclass = sorted(row['product'].split(';'))
    for pc in pclass:
        testcatD[pc].append(row['name'])
    if len(pclass) > 1:
        # print(pclass)
        combD[';'.join(pclass)].append(row['name'])
        
```

```python
combD.keys()
# 好好笑：'保溫杯;冷氣/冷暖空調', '啤酒;豆奶/豆乳',  '妝前乳/隔離霜;豆奶/豆乳'...
```

## Eyeballing

```python
# 每一類別random 2 個出來
# categories = list(catD.keys())
import random
ks, vs = [], []
for k, v in testcatD.items():
    if len(v) == 0: randv = 'No item predicted'
    else:
        idx = random.randint(0, len(v)-1)
        randv = v[idx]
    ks.append(k); vs.append(randv)
category_rand = pd.DataFrame({'category': ks, 'random instance': vs})
```

```python
pd.set_option('display.max_rows', 250)
pd.set_option('display.min_rows', 220)


category_rand.to_csv(f'{root}/predictions/test_predictions_eyeballing1.csv', index = False)
category_rand
```

```python
# (2, 3)多標籤 每一類random 1個出來
col1, col2, col3, col4 = [], [], [], []
for comb, comb_examples in combD.items():
    idx = random.randint(0, len(comb_examples)-1)
    comb_example = comb_examples[idx]
    col1.append(len(comb.split(';')))
    col2.append(comb)
    col3.append(comb_example)
    col4.append(len(comb_examples))
combdf = pd.DataFrame({'category_num':col1, 
                       'category_combination':col2, 
                       'random_instance': col3, 
                       '# of instances':col4})
```

```python
combdf = combdf.sort_values(['category_num', '# of instances'], ascending=[True, False])
combdf.to_csv(f'{root}/predictions/test_predictions_eyeballing2.csv', index = False)
combdf
```
